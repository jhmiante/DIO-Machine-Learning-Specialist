# K-Means

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python

# 1. Gerar dados sintéticos
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
```

```python
# 2. Criar e treinar o KMeans
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)
```

```python
# 3. Obter os rótulos previstos e centros dos clusters
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_
```

```python
# 4. Visualizar clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label="Centroides")
plt.title("Exemplo de KMeans com 4 Clusters")
plt.legend()
plt.show()
```
